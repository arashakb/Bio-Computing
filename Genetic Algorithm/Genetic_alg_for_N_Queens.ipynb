The main problem here is to solve the N-Queens problem using genetic algorithm. In order to make our algorithm more efficient, local search technique also got implmeneted. 

## Some Useful Libraries 

```python
import numpy as np
import random
import time
```

## Permutation Functions

```python
class Permutation:
    def getNum(self, v) :
        n = len(v)
        index = random.randint(0, n - 1)
        num = v[index]
        v[index], v[n - 1] = v[n - 1], v[index]
        v.pop()
        return num

    def generateRandom(self, n) :
        v = [0] * n
        for i in range(n) :
            v[i] = i + 1
        perm = []
        while (len(v)) :  
            perm.append(self.getNum(v))
        return perm
```

## Main Functions

```python
class GA:
    # FITNESS FUNCTION FOR EVALUATION OF SAMPLES
    def fitness(self,chromosome):
        res=0
        n = len(chromosome)
        for i in range(n):
            q1 = chromosome[i]
            for j in range(1,n-i):
                q2 = chromosome[i+j]
                if(abs(q2-q1)==j):res+=1
        return res
    # DEFINING THE CROSSOVER FUNCTION FOR GENERATING NEW PERMUTATIONS
    def order_recombination(self,arr1, arr2, n):
        # print(arr1,arr2)
        p1 = random.randint(0,n-2)
        p2 = random.randint(0,n-2)
        ans = [0] * n
        if p1 > p2:
            tmp = p2    
            p2 = p1
            p1 = tmp
        # print(p1, p2)
        for i in range(p1, p2+1):
            ans[i] = arr1[i]
        # print(ans)

        ans_cnt = p2+1
        arr2_cnt = ans_cnt
        while ans_cnt != p1:
            if arr2[arr2_cnt] not in arr1[p1:p2+1]:
                ans[ans_cnt] = arr2[arr2_cnt]
                # print(ans)
                arr2_cnt+=1
                if arr2_cnt == len(arr2):
                    arr2_cnt = 0
                ans_cnt+=1
                if ans_cnt == len(arr2):
                    ans_cnt = 0
            else:
                arr2_cnt+=1
                if arr2_cnt == len(arr2):
                    arr2_cnt = 0
        return ans
    # MUTATION FUNCTION FOR ADDING STOCHASTICITY
    def mutate(self,x): 
        n = len(x)
        c = random.randint(0, n - 1)
        m = random.randint(0, n - 1)
        tmp = x[c]
        x[c] = x[m]
        x[m] = tmp
        return x
    # GENERATING NEW SAMPLES USING CROSSOVER
    def generating_section(self,population, n, cross_ratio, selection_ratio):
        cross_population = []
        
        for _ in range(cross_ratio):
            a, b = random.sample(population[:selection_ratio], 2)
            cross_population.append(self.order_recombination(a, b, n))
        
        return cross_population
    # GENERATING THE MUTATED CHROMOSOME
    def mutation_section(self,population, mutation_ratio, selection_ratio):
        mutation_population = []
        
        for _ in range(mutation_ratio):
            chrom = random.choice(population[:selection_ratio])
            mutation_population.append(self.mutate(chrom))
        
        return mutation_population
    # MAIN CODE OF THE GENETIC ALGORITHM
    def main_part(self,population, size, temp):
        priority_ratio = 9
        crossover_ratio = 10
        mutation_ratio = 1
        
        population = sorted(population, key=self.fitness)[:priority_ratio]

        population += self.generating_section(population, n, crossover_ratio, priority_ratio) + self.mutation_section(population, mutation_ratio, priority_ratio)
        population = sorted(population, key=self.fitness)[:size]
        
        for i, chrom in enumerate(population):
            N = len(chrom)
            ll = self.fitness(chrom)
            
            for _ in range(3):
                first, second = random.sample(range(N), 2)
                final_chromosome = chrom[:]
                tmp = final_chromosome[first]
                final_chromosome[first] = final_chromosome[second]
                final_chromosome[second] = tmp
                chrom_hard_copy = final_chromosome
                nn = self.fitness(chrom_hard_copy)
                
                if (ll - nn) >= 0 or random.random() < np.exp(ll - nn/temp):
                    chrom = chrom_hard_copy
                    ll = nn
                else: pass
            
            population[i] = chrom
        
        return sorted(population, key=self.fitness)
    # A STANDARD WAY FOR CHROMOSOME REPRESENTATION
    def print_chromosome(self,chrom):
      print("Chromosome = {},  Fitness = {}"
          .format(str(chrom), self.fitness(chrom)))
```

```python
population_size = 20
n = int(input("Enter Number of Queens: ")) 
temp=1
convergence_ratio = 0.99

perm = Permutation()
algorithm = GA()
first = time.time()
population = [perm.generateRandom(n) for _ in range(population_size)]
while True:
    temp *= convergence_ratio
    population = algorithm.main_part(population, population_size,temp)
    
    if algorithm.fitness(population[0]) == 0:
      break
last = time.time()
    

chrom_out = []
# print("Solved in Generation {}!".format(generation-1))
for chrom in population:
    if algorithm.fitness(chrom) == 0:
        print("");
        print("One of the solutions: ")
        chrom_out = chrom
        algorithm.print_chromosome(chrom)
        break
        
board = []

for x in range(n):
    board.append(["x"] * n)
    
for i in range(n):
    board[n-chrom_out[i]][i]="Q"
        

def print_board(board):
    for row in board:
        print (" ".join(row))
        
print()
print_board(board)
print(f'Time of computation: {last-first}')
```
