```python
import numpy as np
import random
import time
from math import sqrt
```

```python
def getNum(v) :
    n = len(v)
    index = random.randint(0, n - 1)
    num = v[index]
    v[index], v[n - 1] = v[n - 1], v[index]
    v.pop()
    return num

def generateRandom(n) :
    v = [0] * n
    for i in range(n) :
        v[i] = i
    perm = []
    while (len(v)) :  
        perm.append(getNum(v))
    return perm
```

```python
def order_recombination(arr1, arr2, n):
    # print(arr1,arr2)
    p1 = random.randint(0,n-2)
    p2 = random.randint(0,n-2)
    ans = [0] * n
    if p1 > p2:
        tmp = p2    
        p2 = p1
        p1 = tmp
    # print(p1, p2)
    for i in range(p1, p2+1):
        ans[i] = arr1[i]
    # print(ans)

    ans_cnt = p2+1
    arr2_cnt = ans_cnt
    while ans_cnt != p1:
        if arr2[arr2_cnt] not in arr1[p1:p2+1]:
            ans[ans_cnt] = arr2[arr2_cnt]
            # print(ans)
            arr2_cnt+=1
            if arr2_cnt == len(arr2):
                arr2_cnt = 0
            ans_cnt+=1
            if ans_cnt == len(arr2):
                ans_cnt = 0
        else:
            arr2_cnt+=1
            if arr2_cnt == len(arr2):
                arr2_cnt = 0
    return ans
# MUTATION FUNCTION FOR ADDING STOCHASTICITY
def mutate(x): 
    n = len(x)
    c = random.randint(0, n - 1)
    m = random.randint(0, n - 1)
    tmp = x[c]
    x[c] = x[m]
    x[m] = tmp
    return x
# GENERATING NEW SAMPLES USING CROSSOVER
def generating_section(population, n, cross_ratio, selection_ratio):
    cross_population = []
    
    for _ in range(cross_ratio):
        a, b = random.sample(population[:selection_ratio], 2)
        cross_population.append(order_recombination(a, b, n))
    
    return cross_population
# GENERATING THE MUTATED CHROMOSOME
def mutation_section(population, mutation_ratio, selection_ratio):
    mutation_population = []
    
    for _ in range(mutation_ratio):
        chrom = random.choice(population[:selection_ratio])
        mutation_population.append(mutate(chrom))
    
    return mutation_population

def main_part(population, size, temp):
        priority_ratio = 40
        crossover_ratio = 50
        mutation_ratio = 10
        
        population = sorted(population, key=fitness)[:priority_ratio]

        population += generating_section(population, n, crossover_ratio, priority_ratio) + mutation_section(population, mutation_ratio, priority_ratio)
        population = sorted(population, key=fitness)[:size]
        
        for i, chrom in enumerate(population):
            N = len(chrom)
            ll = fitness(chrom)
            
            for _ in range(3):
                first, second = random.sample(range(N), 2)
                final_chromosome = chrom[:]
                tmp = final_chromosome[first]
                final_chromosome[first] = final_chromosome[second]
                final_chromosome[second] = tmp
                chrom_hard_copy = final_chromosome
                nn = fitness(chrom_hard_copy)
                
                if (ll - nn) >= 0 or random.random() < np.exp(ll - nn/temp):
                    chrom = chrom_hard_copy
                    ll = nn
                else: pass
            
            population[i] = chrom
        
        return sorted(population, key=fitness)

def print_chromosome(chrom):
  print("Fitness = {}, Chromosome = {}"
      .format(fitness(chrom), str(chrom)))

def fitness(chromosome):
    path=0
    for i in range(len(matrix)):
        gene1 = chromosome[i]
        if(i+1<len(matrix)):
          gene2 = chromosome[i+1]
        else:
          gene2 = chromosome[0]
        path+=matrix[gene1][gene2]
    return path
```
